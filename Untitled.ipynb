```python
from models.mixnet import MixNet
from torch.nn import BCEWithLogitsLoss
from torch.optim import Adam
from torch.optim.lr_scheduler import ExponentialLR


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = MixNet(arch="l", num_classes=1)
model.to(device)
criterion = BCEWithLogitsLoss()
optimizer = Adam(model.parameters(), lr=CFG["lr"])
lr_scheduler = ExponentialLR(optimizer, gamma=0.1, verbose=True)
```

```python
params = sum(p.numel() for p in model.parameters())
params
```

```python
def train_one_epoch(model, optimizer, train_loader, val_loader, criterion, epoch, device):
    model.train()
    train_loss = 0.0
    for i, batch in enumerate(train_loader):
        images, targets = batch

        out = model(images.to(device))
        loss = criterion(out, targets.to(device))

        optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.)
        optimizer.step()

        train_loss += loss.item()

        if i % 500 == 499:
            print(f"Epoch - {epoch}, iter - {i + 1}/{len(train_loader)}: train_loss = {train_loss / 50:.3f}")
            train_loss = 0.0
            break

    val_loss = 0.0
    model.eval()
    with torch.no_grad():
        for i, batch in enumerate(val_loader):
            images, targets = batch

            out = model(images.to(device))
            loss = criterion(out, targets.to(device))
            val_loss += loss.item()
            if i % 1000 == 999:
                print(f"Epoch - {epoch}: val_loss = {val_loss / len(val_loader):.3f}")
                break

    return val_loss / len(val_loader)
```

```python
best_val_loss = np.inf

for epoch in range(1):
    val_loss = train_one_epoch(model, optimizer, train_loader, 
                               val_loader, criterion, epoch, device)
    if epoch % 500 == 499:
        lr_scheduler.step()
    if best_val_loss > val_loss:
        best_val_loss = val_loss
        torch.save(model.state_dict(), f"dist_model_val{best_val_loss:.2f}.pth")
```
